```python
import pandas as pd
from datetime import datetime

# === Load the CSV file ===
file_path = "leading_indicators_from_lseg.csv"  # adjust path if needed
df = pd.read_csv(file_path)

# === Define mapping from EcoSeriesID to readable names ===
id_to_name = {
    202813: "PMI_Services_Business_Activity_US",
    202812: "PMI_Services_Business_Activity_Global",
    202811: "PMI_Manufacturing_Global",
    134896: "Consumer_Confidence_Index",
    201322: "Consumer_Confidence_Expectations",
    135480: "Unemployment_Initial_Claims",
    2024344: "Retail_Sales_Core",
    202587: "Building_Permits_Contribution",
    202482: "10Y_Yield_Constant_Maturity",
    135187: "20Y_Yield_Constant_Maturity"
}

# === Convert date column to datetime ===
df['Date'] = pd.to_datetime(df['PeriodDate'], format="%m/%d/%Y")

# === Map indicator names ===
df['Indicator'] = df['EcoSeriesID'].map(id_to_name)

# === Drop rows with unmapped indicators or missing values ===
df = df[['Date', 'Indicator', 'Series_Value']].dropna()
df = df[df['Indicator'].notnull()]
df = df.rename(columns={'Series_Value': 'Value'})

# === Aggregate duplicates by taking the mean ===
df_agg = df.groupby(['Date', 'Indicator'], as_index=False)['Value'].mean()

# === Pivot to wide format ===
df_wide = df_agg.pivot(index='Date', columns='Indicator', values='Value')

# === Sort by date and save to CSV (optional) ===
df_wide = df_wide.sort_index()
df_wide.to_csv("wide_format_leading_indicators.csv")

# === Preview ===
print(df_wide.tail())
```
