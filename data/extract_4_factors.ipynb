```python
import pandas as pd
from functools import reduce

# === Path to the Excel file ===
file_path = "Quality_Minus_Junk_Factors_Daily.xlsx"  # update this if needed

# === Sheet names and column rename mapping ===
sheets_info = {
    "QMJ Factors": "QMJ",
    "MKT": "MKT-Rf",
    "SMB": "SMB",
    "HML FF": "HML",
    "UMD": "UMD"
}

dfs = []

# === Read and process each sheet ===
for sheet, factor_name in sheets_info.items():
    # Read column A and Z, starting from row 20 (index=19), no headers
    df = pd.read_excel(file_path, sheet_name=sheet, usecols="A,Z", skiprows=19, header=None)
    df.columns = ["Date", factor_name]
    dfs.append(df)

# === Merge all dataframes on Date ===
df_merged = reduce(lambda left, right: pd.merge(left, right, on="Date", how="outer"), dfs)

# ✅ Convert Date column to datetime before sorting
df_merged['Date'] = pd.to_datetime(df_merged['Date'])

# === Sort and forward fill ===
df_merged = df_merged.sort_values("Date")


# === Save to CSV ===
df_merged.to_csv("merged_factors.csv", index=False)

# === Optional preview ===
print(df_merged.tail())

```
